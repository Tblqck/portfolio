```python
import pandas as pd
import hvplot.pandas
import numpy as np
from IPython.display import display
import holoviews as hv
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the CSV file
file_path = r'C:\Users\T blqck\projects\India_rainfall_act_dep_1901_2016_1.csv'
data = pd.read_csv(file_path)

# Preview the columns
print(data.head())

# Display the column names
print(data.columns)
```

```python
We will obtain a line graph that represents the trend of average rainfall over the years.
This visualization can help you identify any patterns or trends in the rainfall data, such as increasing, decreasing, or stable rainfall levels over time.
```

```python

# Extract the relevant columns for analysis
years = data['YEAR']
rainfall_jun_sept = data['Actual Rainfall: JUN-SEPT']

# Plot the trend
plt.figure(figsize=(10, 6))
plt.plot(years, rainfall_jun_sept)
plt.xlabel('Year')
plt.ylabel('Rainfall (in mm)')
plt.title('Rainfall Trend: 1901-2016')
plt.grid(True)
plt.show()



# Extract the columns for each month's rainfall
rainfall_jun = data['Actual Rainfall: JUN']
rainfall_jul = data['Actual Rainfall: JUL']
rainfall_aug = data['Actual Rainfall: AUG']
rainfall_sep = data['Actual Rainfall: SEPT']

# Plotting the trend of monthly rainfall over the years
plt.figure(figsize=(10, 6))
plt.plot(data['YEAR'], rainfall_jun, label='June')
plt.plot(data['YEAR'], rainfall_jul, label='July')
plt.plot(data['YEAR'], rainfall_aug, label='August')
plt.plot(data['YEAR'], rainfall_sep, label='September')
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')
plt.title('Monthly Rainfall Variation: 1901-2016')
plt.legend()
plt.grid(True)
plt.show()
```

```python
Let's focous on the individual months or seasons (June, July, August, September) to determine any specific patterns or trends.
Analyze the average rainfall for each month/season and look for any significant variations or anomalies.
```

```python
# Seasonal Analysis
def perform_seasonal_analysis(data):
    # Calculate the average rainfall for each month/season
    average_rainfall_jun = data['Actual Rainfall: JUN'].mean()
    average_rainfall_jul = data['Actual Rainfall: JUL'].mean()
    average_rainfall_aug = data['Actual Rainfall: AUG'].mean()
    average_rainfall_sep = data['Actual Rainfall: SEPT'].mean()

    # Print the average rainfall for each month/season
    print("Average Rainfall (in mm):")
    print("June:", average_rainfall_jun)
    print("July:", average_rainfall_jul)
    print("August:", average_rainfall_aug)
    print("September:", average_rainfall_sep)

    # Visualize the average rainfall for each month/season
    months = ['June', 'July', 'August', 'September']
    average_rainfall = [average_rainfall_jun, average_rainfall_jul, average_rainfall_aug, average_rainfall_sep]

    plt.figure(figsize=(10, 6))
    plt.bar(months, average_rainfall)
    plt.xlabel('Month/Season')
    plt.ylabel('Average Rainfall (mm)')
    plt.title('Average Rainfall for Each Month/Season')
    plt.grid(True)
    plt.show()

# Calculate Total Annual Rainfall
def calculate_total_annual_rainfall(data):
    # Calculate the total annual rainfall
    data['Total Annual Rainfall'] = data[['Actual Rainfall: JUN', 'Actual Rainfall: JUL',
                                          'Actual Rainfall: AUG', 'Actual Rainfall: SEPT']].sum(axis=1)

    # Print the total annual rainfall for each year
    print("Total Annual Rainfall (in mm):")
    print(data[['YEAR', 'Total Annual Rainfall']])

    # Visualize the total annual rainfall
    plt.figure(figsize=(10, 6))
    plt.plot(data['YEAR'], data['Total Annual Rainfall'])
    plt.xlabel('Year')
    plt.ylabel('Total Annual Rainfall (mm)')
    plt.title('Total Annual Rainfall: 1901-2016')
    plt.grid(True)
    plt.show()



# Perform seasonal analysis
perform_seasonal_analysis(data)

# Calculate total annual rainfall and visualize it
calculate_total_annual_rainfall(data)
```

```python
These average rainfall values provide an understanding of the typical rainfall patterns during the monsoon season in the given dataset.
They highlight the months when rainfall is generally higher (July and August) and when it starts to decrease (June and September). Analyzing these patterns can help in understanding the seasonal variations and planning activities that depend on rainfall, such as agriculture or water resource management.
```

```python
# Extract the columns for June and other months' rainfall
rainfall_jun = data['Actual Rainfall: JUN']
rainfall_jul = data['Actual Rainfall: JUL']
rainfall_aug = data['Actual Rainfall: AUG']
rainfall_sep = data['Actual Rainfall: SEPT']

# Create a DataFrame for correlation analysis
correlation_data = pd.DataFrame({
    'June': rainfall_jun,
    'July': rainfall_jul,
    'August': rainfall_aug,
    'September': rainfall_sep
})

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Get the correlation coefficients with June rainfall
correlation_with_june = correlation_matrix['June']

# Print the correlation coefficients with June rainfall
print("\nCorrelation Coefficients with June Rainfall:")
print(correlation_with_june)
```

```python

# Plot the correlation matrix heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Plot the correlation coefficients with June rainfall
plt.figure(figsize=(8, 6))
correlation_with_june.plot(kind='bar', color='skyblue')
plt.xlabel('Months')
plt.ylabel('Correlation Coefficient')
plt.title('Correlation with June Rainfall')
plt.xticks(rotation=0)
plt.show()
```

```python
Let's discuss the findings better:
```

```python
Correlation Matrix:

June and July: The correlation coefficient between June and July rainfall is approximately 0.009. This value is close to zero, indicating a very weak or no correlation between these two months' rainfall.
June and August: The correlation coefficient between June and August rainfall is around 0.067. Again, this value is close to zero, suggesting a weak or negligible correlation between June and August rainfall.
June and September: The correlation coefficient between June and September rainfall is approximately -0.040. This value is also close to zero, indicating a weak or negligible correlation between June and September rainfall.
July and August: The correlation coefficient between July and August rainfall is about 0.203. This value indicates a weak positive correlation between these two months' rainfall.
July and September: The correlation coefficient between July and September rainfall is approximately 0.335. This value suggests a moderate positive correlation between July and September rainfall.
August and September: The correlation coefficient between August and September rainfall is around 0.317. This value also indicates a moderate positive correlation between August and September rainfall.
Correlation Coefficients with June Rainfall:

The correlation coefficient between June and itself is 1. This value is expected since it represents the correlation of a variable with itself.
The correlation coefficients between June and the other months (July, August, and September) are close to zero, indicating weak or negligible correlations.

Conclusions:
Based on the correlation analysis, we can draw the following conclusions:

There is no significant correlation between June rainfall and the rainfall in the subsequent months (July, August, and September). This suggests that the rainfall in June does not strongly predict or influence the rainfall in the following months.
There is a weak positive correlation between July and August rainfall, indicating a slight tendency for these two months to have similar rainfall patterns.
There is a moderate positive correlation between July and September rainfall, as well as between August and September rainfall. This suggests a somewhat stronger relationship between rainfall in July and September, and between rainfall in August and September.
Overall, the correlation analysis helps us understand the degree of association between different months' rainfall. However, it's important to note that correlation does not imply causation. Additional factors and variables may influence rainfall patterns, and further analysis would be required to determine the underlying causes of these correlations.
```

```python
Let's see Correlation between rainfall and year. 
This might provide insights into any long-term trends or patterns in rainfall over the years.
```

```python
# Calculate the total annual rainfall
total_rainfall = data['Actual Rainfall: JUN-SEPT']

# Calculate the correlation coefficient between rainfall and the year
correlation_coefficient = np.corrcoef(data['YEAR'], total_rainfall)[0, 1]

# Print the correlation coefficient
print("Correlation Coefficient between Rainfall and Year:", correlation_coefficient)
```

```python
Let's Visualize this 
```

```python
# Create a scatter plot of rainfall and year
plt.figure(figsize=(10, 6))
plt.scatter(data['YEAR'], total_rainfall, color='b', alpha=0.5)
plt.xlabel('Year')
plt.ylabel('Total Annual Rainfall (mm)')
plt.title('Scatter Plot: Rainfall vs. Year')
plt.grid(True)
plt.show()



# Create a line plot of rainfall trend over the years
plt.figure(figsize=(10, 6))
plt.plot(data['YEAR'], total_rainfall, color='b')
plt.xlabel('Year')
plt.ylabel('Total Annual Rainfall (mm)')
plt.title('Rainfall Trend: 1901-2016')
plt.grid(True)
plt.show()
```
